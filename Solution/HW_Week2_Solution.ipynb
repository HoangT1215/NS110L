In physics, it often useful to take advantage of mathematical approximations. Demon-
strate for yourself that for “small” angles, θ, the following approximation holds:
$$\sin \theta \approx \tan \theta \approx \theta$$

### Problem 1a
Create a diagram (involving a triangle) that demonstrates why this approxima-
tion is reasonable. Explain the diagram.

### Solution
Constructing a right triangle with a very small acute angle. We can see that one side's length is approximately equal to the length of the hypotenuse, and thus we have $\sin \theta \approx \tan \theta$.

<img src="https://i.stack.imgur.com/UdlyK.gif">

From the diagram, we can see that the triangle ABC has an area of $0.5 \sin{x}$, the blue-colored area is $0.5x$, and the triangle ABD has an area of $0.5 \tan{x}$. We have:

$$0.5 \tan{x} \geq 0.5 x \geq 0.5 \sin{x}$$

$$\Leftrightarrow \tan{x} \geq x \geq \sin{x}$$

$$\Leftrightarrow \cos{x} \leq \frac{\sin{x}}{x} \leq 1$$

Hence, if we let $x \rightarrow 0$, we have $\frac{\sin{x}}{x} \rightarrow 1$. Hence, we have $\sin \theta \approx \theta$ for very small $\theta$. As we already have $\sin \theta \approx \tan \theta$ (when $\theta$ is very small $C \rightarrow D$), we have:

$$\sin \theta \approx \tan \theta \approx \theta$$

# Problem 2
Waves that can be described by a sinusoidal variation are called harmonic waves. A
travelling harmonic wave is mathematically described by writing an expression for the
vertical displacement of the oscillation (y) as a function of both the horizontal position
(x) and time (t) as,

$$y_1(x, t) = A \sin (kx − ωt)$$ (2)

This describes, for example, a wave travelling on a string or rope. For this exercise,
we’ll set the wave parameters to 1, A = k = ω = 1, to simplify the expression:
$$y_1(x, t) = \sin(x − t)$$

```python
# 2a: Plotting traveling wave

import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

# parameters
a = 1
k = 1
w = 1

# space time
x0 = np.linspace(0,2*np.pi,2000)

def y(x,t):
    return a*np.sin(k*x - w*t)

plt.title('Plot of travelling waves')
plt.xlabel('x')
plt.ylabel('y')
for t in range(1,7):
    plt.legend(['t = 1','t = 2','t = 3','t = 4','t = 5','t = 6'])
    plt.plot(x0,y(x0,t*np.pi/6))
    
```

### Problem 2a
As we can see, these waves are called travelling waves as they are travelling forward in time. The wave amplitude is $|y_1(x, t)| = |A \sin (kx − ωt)| \leq A$. The wavelength is $\lambda = \frac{2\pi}{k}$ since we have:

$$y_1(x+\lambda,t) = A \sin (k(x+\frac{2\pi}{k}) − ωt) = A \sin (kx + 2\pi − ωt) = A \sin (kx − ωt) = y_1(x,t)$$

We can also infer the wavelength by looking at the distance between adjacent peaks of the graph, which is exactly $\lambda$.

For the period, we have the period is $T = \frac{2\pi}{\omega}$. We can obtain this by setting $x = 0$ and note that:

$$\sin(-\omega t) = \sin(-\omega t + 2\pi) = \sin(-\omega (t + \frac{2\pi}{\omega}))$$

To find the frequency, we simply take the inverse of wave period:

$$f = \frac{1}{T} = \frac{1}{\frac{2\pi}{\omega}} = \frac{\omega}{2\pi}$$

### Problem 2b
Prove that $$\sin(x-t) + \sin(x+t) = 2\sin(x)\cos(t)$$

### Solution
We recall the following trigonometry identity:

$$\sin(a+b) = \sin a \cos b + \sin b \cos a$$

Similarly, we obtain:

$$\sin(a-b) = \sin a \cos (-b) + \sin (-b) \cos a = \sin a \cos b - \sin b \cos a$$

Due to $\cos (-b) = \cos b$ and $\sin (-b) = -\sin b$.

Return to our problem, we have:
$$\sin(x+t) + \sin(x-t)$$
$$= (\sin x \cos t + \sin t \cos x) + (\sin x \cos (-t) + \sin (-t) \cos x)$$
$$= (\sin x \cos t + \sin t \cos x) + (\sin x \cos t - \sin t \cos x)$$
$$= 2\sin x \cos t$$
Hence, proved.

The peak amplitude of the wave oscillations at any point in space is constant with time, and the oscillations at different points throughout the wave are in phase. As we can see, the waves always peak at $\frac{\pi}{2} + k\pi$.

# Problem 3
In a 100-m race, Lin and Nisha cross the finish line in a dead heat, both taking 10.2 s.
Accelerating uniformly, Lin took 2 s and Nisha 3 s to attain maximum speed, which
they maintained for the rest of the race.

### Problem 3a
What was the acceleration of each sprinter before they hit their steady pace?

### Solution
First, we write the motion equation for both Nisha and Lin. We have the motion equation for Nisha is:

$$x_N = N_0 + v_Nt + 0.5a_Nt^2$$

And the motion equation for Lin is:

$$x_L = L_0 + v_Lt + 0.5a_Lt^2$$

Where $N_0, L_0$ are the initial positions of Nisha and Lin, $v_N, v_L$ are the initial velocities of Nisha and Lin and $a_N, a_L$ are the acceleration of Nisha and Lin respectively.

The max speeds that Nisha and Lin achieved are:

$$m_N = v_N + a_Nt = 0 + a_N\times 2 = 2a_N$$

$$m_L = v_L + a_Lt = 0 + a_L\times 3 = 3a_L$$

Since both reached 100m at 10.2 seconds, we have the following equations:

$$0.5a_N\times 2^2 + m_N(10.2-2)=0.5a_L\times 3^2 + m_L(10.2-3)=100m$$

$$\Leftrightarrow 2a_N + 2a_N(10.2-2)=4.5a_L\times + 3a_L(10.2-3)=100m$$

$$\Leftrightarrow 18.4a_N = 26.1a_L =100$$

$$\Leftrightarrow a_N = 5.4347826087m/s^2 \land a_L = 3.83141762452m/s^2$$

### Problem 3b
What were their respective maximum speeds?

### Solution
Their respective maximum speeds are:

$$m_N = 2a_N = 10.8695652174m/s$$

$$m_L = 3a_L = 11.4942528736m/s$$

### Problem 3c
Which sprinter was ahead at the 6 s mark, and by how much?

### Solution
Nisha's location at 6 second mark is:

$$0.5a_N\times 2^2 + m_N(6-2) = 10a_N = 54.347826087m$$

Lin's location at 6 second mark is:

$$0.5a_L\times 3^2 + m_L(6-3) = 13.5a_L = 51.7241379312m$$

And thus Nisha was ahead of Lin by $2.6236881558m$.

### Problem 3d
What were their respective average speeds over the course of the race?

### Solution
They both have the same average speed of $\frac{100}{10.2}=9.80392156863m/s$

Draw three graphs for each sprinter: Position vs Time, Velocity vs Time, and

```python
# 3c: Plotting runners
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

# space time
time = np.linspace(0,10,1001)
N, L = [], []
a_N = 5.4347826087
a_L = 3.83141762452

def Nisha(t):
    if t <= 2:
        return 0.5*a_N*t**2
    else:
        return 0.5*a_N*2**2+2*a_N*(t-2)
    
def Lin(t):
    if t <= 3:
        return 0.5*a_L*t**2
    else:
        return 0.5*a_L*3**2+3*a_L*(t-3)

for x in time:
    N.append(Nisha(x))
    L.append(Lin(x))

# position vs time

plt.title('Position vs Time')
plt.xlabel('Time')
plt.ylabel('Position')
plt.plot(time,N)
plt.plot(time,L)
```

```python
# 3c: Plotting speed
a_N = 5.4347826087
a_L = 3.83141762452
sp_N, sp_L = [], []
time = np.linspace(0,10,1001)

def speed(t):
    return [a_N*t*(t<=2)+2*a_N*(t>2),a_L*t*(t<=3)+3*a_L*(t>3)]

for x in time:
    sp_N.append(speed(x)[0])
    sp_L.append(speed(x)[1])

# position vs time

plt.title('Speed vs Time')
plt.xlabel('Time')
plt.ylabel('Speed')
plt.plot(time,sp_N)
plt.plot(time,sp_L)
```

```python
# 3c: Plotting speed
a_N = 5.4347826087
a_L = 3.83141762452
ac_N, ac_L = [], []
time = np.linspace(0,10,1001)

def acc(t):
    return [a_N*(t<=2)+0*(t>2),a_L*(t<=3)+0*(t>3)]

for x in time:
    ac_N.append(acc(x)[0])
    ac_L.append(acc(x)[1])

# position vs time

plt.title('Acceleration vs Time')
plt.xlabel('Time')
plt.ylabel('Acceleration')
plt.plot(time,ac_N)
plt.plot(time,ac_L)
```

# Problem 4
Can a goalkeeper at his/her goal kick a soccer ball into the opponents goal without
the ball touching the ground? The distance would be about 95 m. A goalkeeper can
give the ball a speed of 30 m/s.

### Solution
We can think about this problem as: given a velocity, what is the furthest distance an object can travel before it lands?

Indeed, given with the initial velocity $v = 30 m/s$ and the angle $\theta$, we shall apply the projectile equations by splitting into two spatial components $(x,y)$ as below:
$$v_x = v\cos \theta$$
$$v_y = v\sin \theta - gt$$
$$v = \sqrt{v_x^2+v_y^2}$$

From the velocity equations, we can write the motion equations for each component as below:
$$x = vt \cos \theta$$
$$y = vt \sin \theta - 0.5gt^2$$

Now, we solve for the time when the ball lands. Assume the ball lands at $y^* = 0$, we have:
$$vt \sin \theta - 0.5gt^2 = 0$$
$$\Leftrightarrow t = 0 \lor t = \frac{2v\sin \theta}{g}$$

Since we want to find $t > 0$, we take $t = \frac{2v\sin \theta}{g}$

And thus, we can calculate the travel distance of the ball when it lands is:
$$x^* = vt \cos \theta = v\frac{2v\sin \theta}{g}\cos \theta = v^2\frac{2\sin \theta \cos \theta}{g} = v^2\frac{\sin 2\theta}{g} \leq \frac{v^2}{g} \approx 90 < 95$$

Thus, no matter how the ball was kicked, it cannot reach 95 m before it lands.

# Problem 5
Two identical balls are thrown off the roof of a 25 m tall building. One is dropped from
rest and then 1 s later, the second is thrown outward with a velocity with horizontal
and downward components of 10 and 15 m/s, respectively.

### Problem 5a
Which ball hits the ground first?

### Solution
Since we only care about which one hits the ground first, we only consider its downward component. Indeed, we have the motion equations of the balls are:
$$y_1 = y_0 + v_1t + 0.5a_1t^2$$
And since the second ball is dropped 1s second later, we have:
$$y_2 = y_0 + v_2(t-1) + 0.5a_2(t-1)^2$$

Where $y_1,y_2$ is the location of the first and second ball respectively.

Since we drop the balls downward, both ball have gravitational acceleration. On the other hand, the first ball is dropped with zero initial velocity while the second one is dropped with an initial velocity of 15m/s. We have $y_0 = 25, v_1 = 0, v_2 = 15$ and $a_1 = a_2 = -g$. Plug back into the equations, we have:

$$y_1 = 25 - 0\times t + 0.5(-g)t^2 = 25 - 0.5gt^2$$

$$y_2 = 25 - 15(t-1) + 0.5(-g)(t-1)^2$$

Solve for $t$, we have $t_1 = 2.26107815823067, t_2 = 2.19844360243096$. Thus, the second hits the ground faster.

```python
# Solve for y1 and y2
from sympy.solvers import solve
from sympy import Symbol

g = 9.78
t = Symbol('t')

y1 = solve(-0.5*g*t**2+25,t)
y2 = solve(25 - 15*(t-1) + 0.5*(-g)*(t-1)**2,t)

print y1
print y2
```

### Problem 5b

With what velocity does each ball hit the ground?

### Solution
Using previous results, we can calculate that the first ball has velocity $v_1 + a_1t = 0 + 10 \times 5 = 50m/s$ when it hits the ground, and the second ball has velocity $v_2 + a_2t = 15 + 10 \times 2.19844360243096 \approx 37m/s$

### Problem 5c
Which ball travels the greater displacement?

### Solution
Since the first ball is dropped directly downward, we have the ball final position on the ground is directly 25m below (assume no bouncing). On the other hand, the second has a horizontal component in its initial velocity, and thus the motion equation for its horizontal component is:
$$x = x_0 + v_0t + 0.5at^2 = 0 + 10t + 0.5 \times 0 \times t^2 = 10\times 2.19844360243096 \approx 22m$$.

Using the Pythagorean theorem, we have the total displacement of the second ball is $\sqrt{25^2+22^2} \approx 33.3m$, which is further than the first one.
